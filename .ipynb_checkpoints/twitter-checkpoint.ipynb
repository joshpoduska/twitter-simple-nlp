### Modify the inputs to the first cell, then run all cells.

```python
#True = I have my own Twitter API keys and tokens saved in Domino per the instructions in the README
#False = I would like to use the sample data saved with the project
have_keys = False

#for simplicity, the code expects no more than three twitter ids
#this list will only be used if have_keys = True
userid_list = ['@joshpoduska', '@poem_exe', '@p0em_bot']

#list of words or symbols you would like to exclude from the spaCy-processed tokens
#these may be needed if the tweets have lots of repetative hashtags or web links
exclude_tokens = ['poet', 'poetry', 'poem', 'mcropoetry', \
                  'https', 'http', '#', '%', '&', 'amp']
```

```python
import spacy
import squarify 
from wordcloud import WordCloud, ImageColorGenerator
import tweepy
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from IPython.display import Image, HTML, display
```

```python
#load spacy
nlp = spacy.load("en_core_web_lg")
```

```python
#get tweets using tweepy api.user_timeline
#get 200 at a time (the max)
#picks up where it left off for the next batch
#twitter only gives ~3200 tweets so max_tweets isn't needed unless you want LT 3200
#note that the way the code it set up the tweets are pulled in batches of 200

def get_tweets(userid, max_tweets=10000):
    
    #Get your Twitter API credentials
    #save them as Domino User Environment Variables to keep them secure
    #https://support.dominodatalab.com/hc/en-us/articles/205536355-Domino-environment-variables
    consumer_key = os.environ["TWITTER_CONSUMER_KEY"]
    consumer_secret = os.environ["TWITTER_CONSUMER_SECRET"]
    access_key = os.environ["TWITTER_ACCESS_TOKEN"]
    access_secret = os.environ["TWITTER_ACCESS_TOKEN_SECRET"]
    
    #http://tweepy.readthedocs.org/en/v3.1.0/getting_started.html#api
    auth = tweepy.OAuthHandler(consumer_key, consumer_secret)
    auth.set_access_token(access_key, access_secret)
    api = tweepy.API(auth)
       
    api_tweet_list = []
    while max_tweets > len(api_tweet_list):
        try:
            #no max_id to start
            if len(api_tweet_list) == 0: 
                #use extended tweet_mode to get full text
                #but it also comes with a much more complicated json structure
                new_tweets = api.user_timeline(id=userid, tweet_mode='extended', count=200)
                if not new_tweets:
                    print('No tweets found')
                    break
                api_tweet_list.extend(new_tweets)
                last_id = new_tweets[-1].id
            #after the first batch, now use max_id to pick up the next batch
            else: 
                new_tweets = api.user_timeline(id=userid, max_id = last_id - 1, tweet_mode='extended', count = 200)
                if not new_tweets:
                    print(len(api_tweet_list), userid, 'tweets found via the tweepy api')
                    return api_tweet_list
                    break
                api_tweet_list.extend(new_tweets)
                last_id = new_tweets[-1].id
            #catch the case where we've hit the max # of tweets
            if max_tweets <= len(api_tweet_list):
                return api_tweet_list
        except tweepy.TweepError as e:
            print('hit an error')
            # depending on TweepError.code, one may want to retry or wait
            # to keep things simple, we will give up on an error
            break
```

```python
#example of what this function returns
f1 = get_tweets('@poem_exe', 1)[0:3]
f1
```

```python
#getting the full text of the tweet is not straightforward
#the json is different depending on the type of tweet

def extract_text(userid, api_tweet_list):
    
    tweet_text_list = []
    i=1
    for tweet in api_tweet_list:
        status_json = tweet._json 
        #logic to check for retweet
        if "extended_tweet" in status_json: 
            text = status_json['extended_tweet']['full_text']
        elif 'retweeted_status' in status_json: 
            text = status_json['retweeted_status']['full_text']
        else: 
            text = status_json['full_text']        
        #build a list of the text and some metadata
        tweet_text_list.append([userid, tweet.id_str, tweet.created_at, text])
        i+=1
    
    print(len(tweet_text_list), userid, 'tweets full text extracted')
    
    return tweet_text_list
```

```python
#example of what this function returns
f2 = extract_text('@poem_exe',f1)
f2
```

```python
#load into a df
#add a column

'''As a special note. I am using pandas dataframes liberally in this notebook. 
If one was analyzing larger Twitter pulls one could run into real memory issues. 
Here I am limited to 3200 tweets on my dev account. numpy arrays or dicitonaries 
might be a better option than dataframes for large corpuses'''

def load_df(userid, tweet_text_list):
    
    #load the df
    df_tweets = pd.DataFrame(tweet_text_list, columns =['username', 'tweetid', 'date', 'tweet'])
    df_tweets['month'] = df_tweets['date'].dt.month
    df_tweets['year'] = df_tweets['date'].dt.year
    
    #concatenate month and year by creating a list and then loading it as a new column
    #I'm sure there is a more direct way to do this - feel free to improve
    i=0
    list_month_year = []
    #loop through each tweet
    #while i < len(df_tweets):
    for i, tweet in df_tweets.iterrows():
        #concatenate the month and year
        if len(str(df_tweets['month'].values[i])) == 1:
               list_month_year.append(str(df_tweets['year'].values[i]) + '/' + '0' + \
                                      str(df_tweets['month'].values[i]))
        else:
               list_month_year.append(str(df_tweets['year'].values[i]) + '/' + \
                                      str(df_tweets['month'].values[i]))
        i += 1
    
    #add list as a new column in the df
    df_tweets['month_year'] = list_month_year
    
    print(len(df_tweets), userid, 'tweets loaded in a df')
    
    return df_tweets    
```

```python
#example of what this function returns
f3 = load_df('@poem_exe',f2)
f3
```

```python
#use spacy to get the tokens and process them, returning a curated df

def process_tokens(userid, df_tweets):
    
    #create a list of tokens and metadata for the token
    tokens = []
    i = 0
    #loop through each tweet
    while i < len(df_tweets):
        #loop through each token in the tweet
        for token in nlp(df_tweets.tweet[i]):
            #use wordcloud stopwords
            #use spacy to lemmatize and strip
            #use spacy to get POS
            #keep the tweet id and dates from the outer loop
            if not token.is_stop and str(token) and \
                    token.pos_ not in (['PUNCT', 'SYM', 'X', 'DET', 'PART', 'SCONJ', 'SPACE']):
                tokens.append([df_tweets['tweetid'][i], df_tweets['date'][i], \
                               df_tweets['month_year'][i], token.lemma_.strip(), token.pos_])
        i += 1
    
    #load the token list into a df
    df_tokens_raw = pd.DataFrame(tokens, columns =['tweetid', 'date', 'month_year', 'token', 'POS'])
    
    #a bit of manual cleanup on the tokens
    #using the list of excluded text from the beginning of the notebook
    mask = []
    for t in df_tokens_raw['token'].astype(str):
        mask.append(not any(x in t for x in exclude_tokens))
        
    df_tokens_filtered = df_tokens_raw[mask]
    
    print(len(df_tokens_filtered), userid, 'tokens loaded in a df after processing')
    
    return df_tokens_filtered
```

```python
#example of what this function returns
f4 = process_tokens('@poem_exe',f3)
f4
```

```python
#create a WordCloud

def graph_WordCloud(userid, df_tokens_filtered, hide_plot=True):
    
    #filter the df for nouns and proper nouns as those are most interesting
    df_wordcloud = df_tokens_filtered[(df_tokens_filtered['POS'] == 'NOUN') | \
                                      (df_tokens_filtered['POS'] == 'PROPN')]
    
    #merge the tokens into one string as needed by the WordCloud function
    text = " ".join(review for review in df_wordcloud.token)
    
    # Generate a word cloud image
    wordcloud = WordCloud(background_color='white', max_words=100,).generate(text)
        
    #save and display if desired
    fig = plt.figure(figsize=(14,4))
    plt.imshow(wordcloud, interpolation='bilinear')
    plt.axis("off")
    plt.title(userid + '\n' + 'WordCloud')
    file = './results/' + userid + '_WordCloud.png'
    plt.savefig(file, bbox_inches="tight")
    
    print(file, 'saved')   
    
    if hide_plot: 
        plt.close(fig)
        
    return file
```

```python
#example of what this function returns
graph_WordCloud('@poem_exe',f4, False)
```

```python
#plot organized by month/year along the x-axis
#show count of mentions of the top topic as bars with the left axis
#show count of tweets as line with the right axis
#list the top and second most used topic in the x-asix labels
#do a whole bunch of data munging to setup the dfs for the final graph

def graph_TopicTrend(userid, df_tweets, df_tokens_filtered, hide_plot=True):
    
    #keep nouns and proper nouns
    #get a count of tokens used by month/year
    df1= df_tokens_filtered[(df_tokens_filtered['POS'] == 'NOUN') | \
                            (df_tokens_filtered['POS'] == 'PROPN')] \
                            .groupby(['token', 'month_year'], as_index=False).token.agg({'count'})
    
    #as_index=False (used above) doesn't seem to work when grouping by two columns
    #so we have to create a new column in order to get 'token' and 'month_year' back as columns
    df1['x'] = df1.index
    
    #next we have to split 'token' and 'month_year' out as they are in one column currently
    #doing this with a list - there is probably a better way - doing this directly in the df somehow
    
    #do it for month_year 
    i=0
    list_month_year = []
    while i < len(df1):
        list_month_year.append(str(df1['x'].values[i][1]))
        i += 1
        
    #do it for token
    i=0
    list_token = []
    while i < len(df1):
        list_token.append(df1['x'].values[i][0])
        i += 1
    
    #load token and month_year and counts into a df
    df2 = pd.DataFrame()
    df2['token'] = list_token
    df2['month_year'] = list_month_year
    df2['count'] = df1['count'].values
    
    #index by month_year to help with aggregations
    df2.set_index('month_year', inplace=True)
    
    #get the max number of mentions by month_year
    df3 = df2.groupby('month_year').agg({'count': 'max'})
    df3.columns = ['max_count']
    
    #merge max number of mentions back in as a new column
    df4 = pd.merge(df3, df2, on='month_year', how='left')
    df4 = df4.sort_values(['month_year', 'count'], ascending=False)
    
    #only keep tokens that equal the max number of mentions for that month_year
    df5 = df4[df4['count'] == df4['max_count']]
    
    #for simplicity, only keep the first token if there is a tie
    df6 = df5.groupby('month_year').first()
    df6.drop(['count'], axis=1, inplace=True)
    
    #merge the df from the previous step that has the top mentions back into 
    #the df with all the tokens
    df7 = pd.merge(df4, df6, on=['month_year', 'token'], how='left', indicator=True)
    
    #get rid of rows that were the top tokens so we can find the second 
    #most mentioned tokens by month_year
    #plotting two tokens is way more informative than just one
    df7 = df7[(df7['_merge'] != 'both')]
    df7.drop(['max_count_y', 'max_count_x', '_merge'], axis=1, inplace=True)
    
    #go through the same process of finding the most mentioned token
    #on this new df that has the top tokens already removed
    #this gives us a df of the #2 tokens
    df8 = df7.groupby('month_year').agg({'count': 'max'})
    df8.columns = ['max_count']
    df9 = pd.merge(df8, df7, on='month_year', how='left')
    df9 = df9.sort_values(['month_year', 'count'], ascending=False)
    df10 = df9[df9['count'] == df9['max_count']].groupby('month_year').first()
    df11 = df10.drop(['max_count', 'count'], axis=1)
    df11.columns = ['token2']
    
    #merge the df of #1 tokens with the df of #2 tokens
    df12 = pd.merge(df6,df11, on='month_year', how='left', indicator=True)
    df12['month_year_'] = df12.index
    df12['month_token'] = df12['token'] + ' + ' + df12['token2'] + ' ' + df12['month_year_'].map(str)
    
    #get the tweet count by month_year from the df_tweets df that was passed in
    df_tweet_count = df_tweets.groupby('month_year').tweet.agg({'count'})

    #make the plot, save, and show if desired
    fig, ax1 = plt.subplots(figsize=(14,4))
    index = np.arange(len(df12))
    plt.bar(index, df12['max_count'])
    plt.ylabel('# of Top Topic Mentions', fontsize=10)
    plt.xticks(index, df12['month_token'], fontsize=10, rotation=90)
    plt.title(userid + '\n' + 'Top Tweet Topics by Month')
    ax2 = ax1.twinx()  # instantiate a second axes that shares the same x-axis
    index = np.arange(len(df_tweet_count['count']))
    color = 'tab:green'
    ax2.set_ylabel('# of Tweets', color=color)  # we already handled the x-label with ax1
    ax2.plot(index, df_tweet_count['count'], color=color)
    ax2.tick_params(axis='y', labelcolor=color)
    file = './results/' + userid + '_Trends.png'
    plt.savefig(file, bbox_inches="tight")
    
    print(file, 'saved')
   
    if hide_plot:
        plt.close(fig)
        
    return file
```

```python
#example of what this function returns
graph_TopicTrend('@poem_exe',f3, f4, False)
```

```python
#get the top n tokens and load them as a spacy doc, ready for further analysis

def top_n_doc(userid, df_tokens_filtered, top_n):
    
    #remove the '-PRON-' token that spacy creates for all pronouns
    #filter the df for the top n tokens
    df_top = df_tokens_filtered[df_tokens_filtered['token'] != '-PRON-'] \
                .groupby(['token'], as_index=False).token.agg({'count'}) \
                .sort_values('count', ascending=False).head(top_n)
    
    #put all the tokens in one string variable
    doc = " ".join(df_top.index.values)
    
    print(userid, 'spaCy doc loaded')
    
    #load that string into a spacy doc
    return nlp(doc)
```

```python
#example of what this function returns
top_n_doc('@poem_exe', f4, 10)
```

### All the output above showed what each function does
### Now we run all the functions for each user account and display comparison plots

```python
'''
Since we are looping through the twitter accounts,
it would be nice to use dynamic variable creation and referencing, 
but this is tough to do in python.
Instead, just store everything in a list where the first level will represent
the accounts we are looping through. Then we can access each as needed.
'''

uid_list = []
full_tweets = []
text_extracts = []
df_raw = []
df_tokens = []
file_WordCloud = []
file_TopicTrend = []
top_doc = []

#if you have your own twitter keys then pull live data
#this flag is set in the first cell
if have_keys:

    #loop through all the twitter accounts from userid_list
    #this list was set in the first cell
    for uid in userid_list:

        #store the userid
        uid_list.append(uid)
        
        #run each function
        #store the return object of those functions
        #notice that functions down the list require output from previous functions

        full_tweets.append(get_tweets(uid_list[-1]))
        text_extracts.append(extract_text(uid_list[-1], full_tweets[-1]))
        df_raw.append(load_df(uid_list[-1], text_extracts[-1]))  
        df_tokens.append(process_tokens(uid_list[-1], df_raw[-1]))
        file_WordCloud.append(graph_WordCloud(uid_list[-1], df_tokens[-1]))
        file_TopicTrend.append(graph_TopicTrend(uid_list[-1], df_raw[-1], df_tokens[-1]))
        top_doc.append(top_n_doc(uid_list[-1], df_tokens[-1], 10))

#if you don't have your own twitter keys then use the sample data stored with this Domino project
else:
    
    #set users and sample data
    uid_list = ['@joshpoduska', '@poem_exe', '@p0em_bot']
    df_raw.append(pd.read_csv('df_@joshpoduska'))
    df_raw.append(pd.read_csv('df_@poem_exe'))
    df_raw.append(pd.read_csv('df_@p0em_bot'))
   
    #loop through each user in the sample data
    i = 0
    while i < 3:
        
        #run each function
        #store the return object of those functions
        #notice that functions down the list require output from previous functions
 
        df_tokens.append(process_tokens(uid_list[i], df_raw[i]))
        file_WordCloud.append(graph_WordCloud(uid_list[i], df_tokens[-1]))
        file_TopicTrend.append(graph_TopicTrend(uid_list[i], df_raw[i], df_tokens[-1]))
        top_doc.append(top_n_doc(uid_list[i], df_tokens[-1], 10))
            
        i += 1
```

```python
#Frequency of POS Usage by Tweeter
#creates a df table where the top POS tweeters
#are highlighted in yellow

i=0
df_POS_comp = []
#loop through the twitter accounts
for uid in userid_list:
    #get the count for each POS - will be the numerator
    df_POS_comp.append(df_tokens[i].groupby('POS').agg({'tweetid': 'count'}) \
                       .sort_values('tweetid', ascending=False))
    #create a column that repeats the total sum of POS usage - will be the demonator
    df_POS_comp[-1]['sum'] = np.repeat(df_POS_comp[-1].agg({'tweetid': 'sum'}).values, len(df_POS_comp[-1]))
    #divide to get the percentage
    df_POS_comp[-1]['perc'] = np.round(100 * df_POS_comp[-1]['tweetid'] / df_POS_comp[-1]['sum'],1)
    #cleanup
    df_POS_comp[-1].drop(['sum', 'tweetid'], axis=1, inplace=True)
    #rename columns
    df_POS_comp[-1].columns = [uid + '_' + 'perc']
    i+=1

#figure out what to display
if len(userid_list) == 1:
    df_POS_perc = df_POS_comp[0]
if len(userid_list) == 2:
    df_POS_perc = pd.merge(pd.merge(df_POS_comp[0], df_POS_comp[1], on='POS'))
if len(userid_list) == 3:
    df_POS_perc = pd.merge(pd.merge(df_POS_comp[0], df_POS_comp[1], on='POS'), df_POS_comp[2], on='POS')

#cool function to highlight max values
#https://pandas.pydata.org/pandas-docs/stable/user_guide/style.html
def highlight_max(s):    
    is_max = s == s.max()
    return ['background-color: yellow' if v else '' for v in is_max]
 
print('Percent of POS Usage by Tweeter')

#transpose the df and apply the highlighting function
df_POS_perc.transpose().style.apply(highlight_max)
```
